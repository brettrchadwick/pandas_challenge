```python
#import modules
import pandas as pd
import numpy as np
```

```python
#read in the purchasedata csv file and convert it to a dataframe, and show first 5 lines to get idea of data 
file='Resources/purchase_data.csv'
orig_df=pd.read_csv(file)
orig_df.head()
```

```python
#find total number of players in data 
unique_players=orig_df['SN'].nunique()
unique_players
```

```python
#find total number of unique items
unique_items=orig_df['Item ID'].nunique()
unique_items
```

```python
#find average purchase price of items
avg_price_item=orig_df['Price'].mean()
avg_price_item
```

```python
#total number of purchases
total_purchase=orig_df['Purchase ID'].count()
total_purchase
```

```python
#total revenue from purchases
total_revenue=orig_df['Price'].sum()
total_revenue
```

```python
#summary dataframe for total players, unique items, avg purchase price, total revenue and total purchases and format columns
summary_df1=pd.DataFrame([{'Number of Players':unique_players, 'Number of Items':unique_items,'Average Purchase Price':avg_price_item,'Total Number of Purchases':total_purchase,'Total Revenue':total_revenue}])
summary_df1['Average Purchase Price']=summary_df1['Average Purchase Price'].map('${:.2f}'.format)
summary_df1['Total Revenue']=summary_df1['Total Revenue'].map('${:.2f}'.format)
summary_df1
```

```python
#getting gender totals and percentage
gender_total_df=orig_df.groupby(['Gender'])
gender_totals=gender_total_df['SN'].nunique()
percent_genders=gender_totals/unique_players*100
```

```python
#create summary df for gender totals and format display output
gender_summary_df=pd.DataFrame({'Number of Players':gender_totals,'Percentage of Players':percent_genders})
gender_summary_df['Percentage of Players']=gender_summary_df['Percentage of Players'].map('{:.2f}%'.format)
gender_summary_df
```

```python
#calculating values for purchase price, average purchase, totals all by gender
gender_purchases=gender_total_df['Purchase ID'].count()
gender_purchase_totals=gender_total_df['Price'].sum()
gender_average_person=gender_purchase_totals/gender_totals
duplicate_total_gender=gender_total_df['SN'].count()
gender_average=gender_purchase_totals/duplicate_total_gender
```

```python
#create summary df for summary analysis by gender and formatting columns for readability
gender_analysis_summary_df=pd.DataFrame({'Purchase Count':gender_purchases, 'Average Purchase Price':gender_average,'Total Purchase Value':gender_purchase_totals,'Average Total Purchase Per Person':gender_average_person})
gender_analysis_summary_df['Average Purchase Price']=gender_analysis_summary_df['Average Purchase Price'].map('${:.2f}'.format)
gender_analysis_summary_df['Total Purchase Value']=gender_analysis_summary_df['Total Purchase Value'].map('${:.2f}'.format)
gender_analysis_summary_df['Average Total Purchase Per Person']=gender_analysis_summary_df['Average Total Purchase Per Person'].map('${:.2f}'.format)

gender_analysis_summary_df
```

```python
#create bins and names for bins for age demographics
bins=[0,9,14,19,24,29,34,39,60]
bin_names=['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']
```

```python
#create a group based off the bins
orig_df['Age Brackets']=pd.cut(orig_df['Age'], bins, labels=bin_names)
orig_df.head()
```

```python
#group by the bin categories
age_df=orig_df.groupby('Age Brackets')
```

```python
#calculating counts and percentages by age brackets
age_totals=age_df['SN'].nunique()
age_percent=age_totals/unique_players*100
```

```python
#create new DF for age demographics and clean data up
age_summary=pd.DataFrame({'Total Count':age_totals, 'Percentage of Players':age_percent})
age_summary['Percentage of Players']=age_summary['Percentage of Players'].map('{:.2f}%'.format)
age_summary
```

```python
#obtain purchase count, avg purchase price, total purchase value and avg purchase per person
age_purchase_count=age_df['Purchase ID'].count()
age_total_purchases=age_df['Price'].sum()
age_duplicate_total=age_df['SN'].count()
age_avg_purchase=age_total_purchases/age_duplicate_total
age_avg_person=age_total_purchases/age_totals
```

```python
#create a summary Df from above calculated values and clean data up
age_analysis_df=pd.DataFrame({'Purchase Count':age_purchase_count,'Average Purchase Price':age_avg_purchase,'Total Purchase Value':age_total_purchases,'Avg Total Purchase Per Person':age_avg_person})
age_analysis_df['Average Purchase Price']=age_analysis_df['Average Purchase Price'].map('${:.2f}'.format)
age_analysis_df['Total Purchase Value']=age_analysis_df['Total Purchase Value'].map('${:.2f}'.format)
age_analysis_df['Avg Total Purchase Per Person']=age_analysis_df['Avg Total Purchase Per Person'].map('${:.2f}'.format)

age_analysis_df
```

```python
#finding Top Spenders values for use in next dataframe
top_spender_df=orig_df.groupby('SN')
spender_purchase_count=top_spender_df['Purchase ID'].count()
spender_total_purchase=top_spender_df['Price'].sum()
spender_avg_purchase=spender_total_purchase/spender_purchase_count
```

```python
#create DataFrame from Top Spenders Values and sort descending by Total Purchase Value
top_spender_analysis=pd.DataFrame({'Purchase Count':spender_purchase_count,'Average Purchase Price':spender_avg_purchase,'Total Purchase Value':spender_total_purchase})
top_spender_analysis=top_spender_analysis.sort_values(['Total Purchase Value'], ascending=False)
top_spender_analysis['Total Purchase Value']=top_spender_analysis['Total Purchase Value'].map('${:.2f}'.format)
top_spender_analysis['Average Purchase Price']=top_spender_analysis['Average Purchase Price'].map('${:.2f}'.format)

top_spender_analysis.head()
```

```python
#finding most popular items
popular_item_df=orig_df.groupby(['Item ID','Item Name'])
pop_purchase_count=popular_item_df['Purchase ID'].count()
pop_item_price=popular_item_df['Price'].mean()
pop_item_total=popular_item_df['Price'].sum()                                
```

```python
#create DataFrame to show most popular items
popular_item_analysis_unsorted=pd.DataFrame({'Purchase Count':pop_purchase_count,'Item Price':pop_item_price,'Total Purchase Value':pop_item_total})
popular_item_analysis=popular_item_analysis_unsorted.sort_values(['Purchase Count'], ascending=False)
popular_item_analysis['Item Price']=popular_item_analysis['Item Price'].map('${:.2f}'.format)
popular_item_analysis['Total Purchase Value']=popular_item_analysis['Total Purchase Value'].map('${:.2f}'.format)
popular_item_analysis.head()
```

```python
#finding most profitable items
profit_item=popular_item_analysis_unsorted.sort_values(['Total Purchase Value'], ascending=False)
profit_item['Total Purchase Value']=profit_item['Total Purchase Value'].map('${:.2f}'.format)
profit_item['Item Price']=profit_item['Item Price'].map('${:.2f}'.format)

profit_item.head()
```
